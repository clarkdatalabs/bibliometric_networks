```python
import csv
import networkx as nx
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
DG = nx.DiGraph()
```

```python
link_dict = {}

with open('links_by_book_without_commentary.csv', "r") as csvfile:
    reader = csv.reader(csvfile)
    
    next(reader)
    for i in reader:
#         print(i)
#         text_referenced = i[0]
#         referencer = i[1]
#         count = i[2]
        link_dict[i[0]] = ({'text': i[1]}, {'weight':i[2]})
        edgeData= {"weight": i[2]}
        DG.add_edge(i[1],i[0], weight=i[2])
        
        
        
# ['Sefer Mitzvot Gadol', 'Tur', '3510']
# ['Daf Shevui to Ketubot', 'Ketubot', '2827']
# ['Bereishit Rabbah', 'Genesis', '2321']
# ['Sefer Mitzvot Gadol', 'Shulchan Arukh, Choshen Mishpat', '2271']
```

```python
layout = nx.spring_layout(DG)
nx.draw_spring(DG)
edgeDataLabels = {}

for a, b in DG.edges():
    # print(DG.get_edge_data(a, b, {"weight":0})["weight"])
    edgeDataLabels[(a, b)] = str(DG.get_edge_data(a, b,["weight"]))
    # retrieve the edge data dictionary
    
```
