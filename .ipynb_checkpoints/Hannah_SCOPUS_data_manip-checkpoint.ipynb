# Bibliographic Networks: A Python Tutorial

Networks can provide significant measures to identify data driven patterns and dependencies. Though, given a data file it can be difficult to discern how one may approach creating such a network. In this tutorial, we will use a bibliographic data file downloaded from a query search in <a href = https://www.scopus.com/search/form.uri>Scopus</a> to walk through the process of cleaning the data file, writing a python script to parse the data into nodes and edges, computing graphical measures using <a href = https://networkx.github.io/documentation/stable/index.html>NetworkX</a>, and creating an interactive network display using <a hred = http://holoviews.org/>HoloViews</a>. 

### Notes on Data Manipulation in Excel

This section is specific to SCOPUS downloaded files, as some downloadable queries are unfortunately imperfect. Firstly, some rows are skewed from inaccurate reading and parsing. Simply scroll through the file and delete any rows where the data is clearly mismatched (i.e. an author name in the 'Title' column, a numerical value in a non-numerical column, etc.). 

Additionally, across several different queries, we discovered duplicates in entry 'Title,' with other columns containing conflicting data. To fix this issue for the purposes of producing a network, we want to remove duplicates. 

In Excel with your csv file open, select Date -> Table Tools -> Remove Duplicates. Indicate that the csv file has headers, as all SCOPUS files will, and only select the 'Title' column by which duplicates will be identified. 

After executing this command, it is important to save the file as a csv. Otherwise, Excel may default to a txt, or other format, and some data features may be lost. By continuously saving the file as a csv, we ensure that it will continue to be compatible with the python code for this tutorial.

Generally, for the case of creating a connected network, we want the rows in our bibliographic data file to have an identifiable title and list of references.

### Scopus and bibliometric specific: fixing the data file 

The downloaded Scopus file identifies a title for a source in each row. The column 'References' indicates a semicolon delimited list of references in MLA/APA format. To make this information useful, we must parse title names from each reference in the list. To understand the function below, look at the formating of the 'References' column in your Scopus file. 

```python
def remove_whitespace(title): 
    if title[0] == " ": 
        title = title[1:]
    if title[-1] == " ": 
        title = title[:-1]
    return title

# Scopus specific: parses a title from an MLA/APA reference
def parse_title(reference): 
    title = reference[0:reference.find(', ,')]
    authors = ""
    while title.find('., ') != -1: 
        authors = authors + title[0:title.find('., ')]
        title = title[title.find('., ') + 2:]
    title = title[0:title.find(',')]
    remove_whitespace(title)
    if title[0] == '(': 
        title = title[title.find(')') + 1:]
    if title[-1] == ')': 
        title = title[0:title.find(')') -1]
    remove_whitespace(title)
    if 'https://' and 'http://' not in title and title != " " and title != "": 
        # only include journal-type references, and do not include empty references
        if title[0] == " ": 
            title = title[1:]
        if title[-1] == " ": 
            title = title[:-1]
        return title
    
references = "Ahmed, H., (2017) Cambridge engineering: the first 150 years, , Third Millennium Publishing, London; Almufti, H., Willford, M., (2013) REDi™ rating system: resilience-based earthquake design initiative for the next generation of buildings, Arup, , http://publications.arup.com; Ambraseys, N., Bilham, R., Corruption kills (2011) Nature, 469, p. p153; (1993) Earthquake hazard and risk in the UK, p. 28. , Arup Department of the Environment, London; Berlin, I., (1953) The hedgehog and the fox: an essay on tolstoy’s view of history, , Weidenfeld & Nicolson, London; Bevere, L., Grollimund, B., (2012) Lessons from recent major earthquakes, , http://media.swissre.com/documents, Swiss Reinsurance Company Ltd, Zurich; Blockley, D., (2012) Engineering: a very short introduction, , Oxford University Press, Oxford; Blockley, D., Godfrey, P., (2017) Doing it differently: systems for rethinking infrastructure, , 2, ICE Publishing, London; Bookstaber, R., (2017) The end of theory: financial crises, the failure of economics and the sweep of human interaction, , Princeton University Press, Princeton; Booth, E., (2014) Earthquake design practice for buildings, , ICE Publishing, London; Borchers, J., Out of the rubble: how can engineers overcome the problems between design and implementation in rural Nepal? (2017) Struct Eng, 95 (11); Bothara, J., Sharpe, R., Seismic protection in developing countries: where are the gaps in our approach? (2003) 7th Pacific conference on earthquake engineering, Christchurch New Zealand. New Zealand Society for Earthquake Engineering, , In:, Wellington; Bruneau, M., Chang, S., Eguchi, R., Lee, G., O’Rourke, T., Reinhorn, A., Shinozuka, M., von Winterfeldt, D.A., Framework to quantitatively assess and enhance the seismic resilience of communities (2003) Earthq Spectra, 19 (4), pp. 733-752; Bruyker, D., The 1980 earthquake in Southern Italy: rescue of trapped victims and mortality (1983) Bull World Health Organ, 612 (6), pp. 1021-1025; Burton, C., Silva, V., Assessing integrated earthquake risk in OpenQuake with an application to Mainland Portugal (2016) Earthq Spectra, 32 (3), pp. 1383-1403; Calvi, M., Sullivan, T., Welch, D., A seismic performance classification framework to provide increased seismic resilience (2014) Perspectives on European earthquake engineering and seismology, 134, pp. 361-400. , Geotechnical, geological and earthquake engineering, In:, Ansal, A, (ed) Springer, Cham; Cannon, T., Disasters, climate change and the significance of ‘culture (2015) Cultures and disasters: understanding cultural framings in disaster risk reduction, , In:, Krüger, F, Bankoff, G, Cannon, T, Orlowski, B, Schipper, ELF, (eds) Routledge, London; (2012) Final report volume 2: the performance of Christchurch CBD buildings, , http://canterbury.royalcommission.govt.nz, Canterbury Earthquakes Royal Commission; (2004) EN 1998: design of structures in earthquake regions, , European Centre for Standardization, Brussels; (2014) Canterbury wellbeing index June 2014, , Canterbury Earthquake Recovery Authority, Christchurch; Clarke, D., Dercon, S., (2016) Dull disasters, , Oxford University Press, Oxford; Comerio, M.C., (2013) Housing recovery in Chile: a qualitative mid-program review. Department of Architecture University of California, Berkeley PEER report 2013/01; Da Silva, J., (2012) Shifting agendas: response to resilience: the role of the engineer in disaster risk reduction. The 2012 Brunel lecture; Despotaki, V., Sousa, L., Burton, C., (2017) Using resilience indicators in the prediction of earthquake recovery, , http://earthquakespectra.org/doi/abs/10.1193/071316EQS107M, Earthq Spectra; Dolce, M., Di Pasquale, G., Speranza, E., Fumagalli, F., A multipurpose method for seismic vulnerability assessment of urban areas (2012) Proceedings of the 15th world conference on earthquake engineering, , In:, Lisbon; (1994) Nepal Building Code NBC 105: seismic design of buildings in Nepal, , Department of Urban Development and Building Control, Kathmandu; (2003) The Kocaeli, Turkey earthquake of 17 August, p. 1999. , https://www.istructe.org, EEFIT, Institution of Structural Engineers, London. from the EEFIT section of; (2013) Recovery two years after the 2011 Tōhoku earthquake and tsunami: a return mission report by EEFIT, , https://www.istructe.org, Institution of Structural Engineers, London. from the EEFIT section of; (2016) M7.8 Gorkha, Nepal earthquake on April 25, 2015 and its aftershocks, , Earthquake Engineering Research Institute, Oakland; (2017) Earthquake early-warning systems finally becoming a reality, , http://www.foxnews.com/tech/2017/02/06/earthquake-early-warning-systems-finally-becoming-reality.html, DOA 10/12/17; (2001) Global earthquake safety initiative (GESI) pilot project, , United Nations Centre for Regional Development, Nagoya; Ghosh, S., Huyck, C.K., Greene, M., Gill, S.P., Bevington, J., Svekla, W., DesRoches, R., Eguchi, R.T., Crowdsourcing for rapid damage assessment: the global earth observation catastrophe assessment network (GEO-CAN) (2011) Earthq Spectra, 27, pp. S179-S198; Gibb, D., (2014) Collectively we are stronger: the 10th Brunel lecture, , https://www.ice.org.uk/eventarchive/10-brunel-international-lecture-cambridge, Institution of Civil Engineers, London; Green, R.A., Unauthorised development and seismic hazard vulnerability: a study of squatters and engineers in Istanbul, Turkey (2008) Disasters, 32 (3), pp. 358-376; Harford, T., (2015) Development needed? Just give cash. Financial Times, , Weekend Magazine, October 17/18; Hewitt, K., Framing disaster in the ‘global village’: cultures of rationality in risk, security and news (2015) In: Krüger F, Bankoff G, Cannon T, Orlowski B, Schipper ELF (eds) Cultures and disasters: understanding cultural framings in disaster risk reduction, , In: Routledge, London; Housner, G., (1989) The second mallet milne lecture: coping with natural disasters: the international decade for natural disaster reduction, , The Society for Earthquake and Civil Engineering Dynamics (SECED), London; (2014) World disasters report: focus on culture and risk International Federation of Red Cross and Red Crescent Societies, , http://www.ifrc.org/world-disasters-report-2014; Iwasaki, N., Role and functions of local communities in earthquake rescue, shelter administration and reconstruction (2000) Int J Jpn Sociol, 9 (1), pp. 111-119; Johnson, L., Mamula-Seadon, L., Transforming governance: how national policies and organizations for managing disaster recovery evolved following the 4 September 210 and 22 February 2011 Canterbury earthquakes (2014) Earthq Spectra, 30 (1), pp. 577-605; Johnson, L., Rabinovici, S., Kang, G., Mahin, S., (2016) California earthquake early warning system benefit study. CSSC Publication 16-04, California Seismic Safety Commission, Sacramento, California; Kahneman, D., (2011) Thinking fast, thinking slow, , Penguin Books, London; Kelly, D., (2012) Quake safety cost must relate to risk. The Press, Christchurch, (15/06/12), Perspective; Kong, Q., Allen, R., Schreier, L., Kwon, Y.-W., MyShake: a smartphone seismic network for earthquake early warning and beyond (2016) Sci Adv, 2 (2); Krüger, F., Bankoff, G., Cannon, T., Orlowski, B., (2015) Cultures and disasters: understanding cultural framings in disaster risk reduction, , (eds), Routledge, London; Langenbach, R., (2009) Don’t tear it down! Preserving the Earthquake Resistant Vernacular Architecture of Kashmir, , UNESCO, New Delhi; (2014) Buildings to be graded for earthquake safety, , http://www.latimes.com, Los Angeles Times Garcetti wants L.ADOA 10/12/17; Lucas, B., Hanson, J., Claxton, G., (2014) Thinking like an engineer: implications for the education system, , Royal Academy of Engineering, London; Makris, N., The dynamics of rocking isolation (2015) SECED Newsl, 26 (2); McNutt, S.R., Seismic monitoring and eruption forecasting of volcanoes: a review of the state-of-the-art and case histories (1996) Monitoring and mitigation of volcano hazards, , Scarpa R, Tilling R, (eds), Springer, Berlin; McVerry, G.H., Gerstenberger, M.C., Rhoades, D.A., Stirling, M.W., Design spectra for the reconstruction of Christchurch (2012) 15th world conference on earthquake engineering, , In:, Lisboa; Mouiller, T., Krimgold, F., (2015) Building regulation for resilience: managing risks for safer cities, , https://www.gfdrr.org/sites/default/files/publication/BRR%20report.pdf, The World Bank, Washington D.C; Muir Wood, R., (2016) The cure for catastrophes, , Oneworld Publications, London; Musson, R.M.W., Fatalities in British earthquakes (2003) Astron Geophys, 44, pp. 14-16. , https://academic.oup.com/astrogeo/article/44/1/1.14/175896/Fatalities-in-British-earthquakes; Musson, R.M.W., (2012) The million death quake, , Pagrave Macmillan, New York; Musson, R.M.W., The eleventh Mallet–Milne lecture: a history of British seismology (2013) Bull Earthq Eng, 11, pp. 715-861; (2015) Letters page, 14th September 2015, , New Civil Engineer; Noji, E.K., Armenian, H.K., Oganessian, A., Issues of rescue and medical care following the 1988 Armenian earthquake (1993) Int J Epidemiol, 22 (6), pp. 1070-1076. , https://academic.oup.com/ije/article-abstract/22/6/1070/650167; Oliver-Smith, A., Conversations in catastrophe (2015) Cultures and disasters: understanding cultural framings in disaster risk reduction, , In:, Krüger, F, Bankoff, G, Cannon, T, Orlowski, B, Schipper, ELF, (eds) Routledge, London; Olson, R.S., Olson, R.A., (1987) Urban heavy rescue. Earthq Spectra, 3 (4), pp. 645-658; Park, R., Paulay, T., (1975) Reinforced concrete structures, , Wiley, New York; Paulay, T., (1993) The fourth Mallet-Milne lecture: simplicity and confidence in seismic design, , SECED/John Wiley, Chichester; Penmellen Boret, S., Shibayama, A., Learning from earthquake memorials: the case of the great East Japan disaster. Paper 1636, 16th world conference on earthquake engineering (2017) Santiago de Chile; Platt, S., (2012) Reconstruction in Chile post 2010 earthquake; report funded by EPSRC, , http://www.carltd.com, Cambridge Architectural Research, ImageCat and the Humanitarian Centre; Platt, S., Factors affecting the speed and quality of post disaster recovery and resilience (2017) Earthquake engineering and structural dynamics in memory of Prof. Ragnar Sigbjörnsson: selected topics. Geotechnical, Geological and Earthquake Enginnering, Springer, , http://www.carltd.com, In:, Rupakhety, R, Olafsson, S, (eds), Netherlands; Platt, S., So, E., Speed or deliberation: a comparison of post-disaster recovery in Japan, Turkey and Chile (2017) Disasters, 41 (4), pp. 696-727; Platt, S., Hughes, M., Brown, D., Measuring resilience and recovery (2015) 7th international conference on seismology and earthquake engineering, , http://www.carltd.com, In:, Tehran; Poggi, V., Durrheim, R., Mavonga, G., Weatherill, G., Gee, R., Pagani, M., Nyblade, A., Delvaux, D., Assessing seismic hazard of the East African rift: a pilot study from GEM and AfricaArray (2017) Bull Earthq Eng, 15 (11), pp. 4499-4529; Press, F., An international decade for natural hazard reduction (1984) Keynote address, 8th world conference on earthquake engineering, , In:, San Francisco; Rayasam, R., (2013) Do more women on the board mean better results? The New Yorker, 19-11-2013, , http://www.newyorker.com; Roberts, D., (2010) A lesson from Haiti: are search and rescue teams worth it?, , http://philanthropyaction.com/nc/a_lesson_from_haiti_are_search_and_rescue_teams_worth_it, Philanthropy action, February; Sanquini, A., Thapaliya, S., Wood, M., Baiocchi, M., Hilley, G., Encouraging earthquake-resistant construction: a randomized controlled trial in Nepal (2016) Earthq Spectra, 32 (4), pp. 1975-1988; Schmidtlein, M., Shafer, J., Berry, M., Cutter, S., Modeled earthquake losses and social vulnerability in Charleston, South Carolina (2011) Appl Geogr, 31 (2011), pp. 269-281; Schumacher, E., (1973) Small is beautiful: a study of economics as if people mattered, , Vintage, London; (2011) SEAONC Rating system for the expected earthquake performance of buildings, , www.seaonc.org, In, SEAOC convention proceedings; Silva, V., Crowley, H., Pagani, M., Monelli, D., Pinho, R., Development of the OpenQuake engine, the global earthquake model’s open-source software for seismic risk assessment (2014) Nat Hazards, 72 (3), pp. 1409-1427; Silver, N., (2012) The signal and the noise: the art and science of prediction, , The Penguin Press, London; Smolka, A., Extreme geohazards: risk management from a (re)insurance perspective (2014) Extreme natural hazards, disaster risks and societal implications, , Ismail-Zadeh A, (ed), Cambridge University Press, Cambridge; Solnit, R., (2009) A paradise built in hell, , Penguin Books, London; Spence, R., The eleventh Mallet–Milne lecture: saving lives in earthquakes: successes and failures since 1960 (2007) Bull Earthq Eng, 5 (2); Stallings, R., Media discourse and the social construction of risk (1990) Soc Probl, 37 (1), pp. 80-95; Stone, H., D’Ayala, D., Wilkinson, S., (2017) The use of emerging technology in post disaster reconnaissance missions, , EEFIT, London; Toquica, M., Burton, C., Valcárcel, J.A., (2017) Measuring social vulnerabilities to earthquakes in the Andean region. In: Proceedings of the 16th world conference on earthquake engineering, Santiago de Chile; Twigg, J., Mosel, I., Emergent groups and spontaneous volunteers in urban disaster response (2017) Environment and Urbanization, 29 (2). , http://journals.sagepub.com/doi/full/10.1177/0956247817721413; (2015) The Sendai framework for disaster risk reduction, 2015–2030, , United Nations Office for Disaster Risk Reduction, Geneva; (2017) Kathmandu, , https://en.wikipedia.org/wiki/Kathmandu, DOA 07-12-2017; (2017) Nepal, , https://en.wikipedia.org/wiki/Nepal, DOA 07-12-2017; (2017) Demographics of New Zealand DOA 07-12-2017, , https://en.wikipedia.org/wiki/Demographics_of_New_Zealand; (2017) Christchurch earthquake, , https://en.wikipedia.org/wiki/2011_Christchurch_earthquake, DOA 07-12-2017; (2017) McLean’s mansion, Christchurch DOA 08-12-2017, , https://en.wikipedia.org/wiki/McLean%27s_Mansion; Prosecutions, , https://en.wikipedia.org/wiki/2009_L%27Aquila_earthquake#Prosecutions, Wikipedia (2017f) L’Aquila earthquakeDOA 07-12-2017; (2017) Critics say Japan ignored warnings of nuclear disaster, , http://www.nytimes.com/2012/03/10/world/asia/critics-say-japan-ignored-warnings-of-nuclear-disaster.html, DOA 07-12-2017; (2017) Smeatonian Society of Civil Engineers, , https://en.wikipedia.org/wiki/Smeatonian_Society_of_Civil_Engineers, DOA 08-12-2017; Worden, C.B., Wald, D.J., Allen, T.I., Lin, K., Garcia, D., Cua, G., A revised ground-motion and intensity interpolation scheme for ShakeMap (2010) Bull Seismol Soc Am, 100 (6), pp. 3083-3096"
refs = references.split('; ')
for ref in refs: 
    print(parse_title(ref))
    print('--')
```

```python
import csv
import networkx as nx
import numpy as np
import pandas as pd

G = nx.Graph() 

with open(file_name) as csvfile:
    reader = csv.DictReader(csvfile)
    # each row contains information for a source
    elist = []
    nlist = {} # key:index, value:title
    tlist = []
    
    index = 0
    e1 = 0
    e2 = 0
    
    for row in reader:
        # add node with unique identifier
        N1 = row['Title']
        if N1 not in tlist:
            tlist.append(N1)
            e1 = index
            nlist[index] = N1
            index = index + 1
        else: 
            # find the e1 index by searching the nlist to find the title 
            print(N1)
            for i, t in nlist.items(): 
                if t == N1: 
                    e1 = i
            
        # add an edge for each source and its references
        refs = row['References'].split(';')
        for ref in refs: 
            # find Title in reference
            title = ref[0:ref.find('(')]
            authors = ""
            while (title.find('.,') != -1): 
                authors = authors + title[0:title.find('.,')]
                title = title[title.find('.,') + 2:]
            N2 = title
            print(title)
            if 'https://' not in title and 'http://' not in title and title != " " and title != "": 
                # only include journal-type references 
                N2 = title
                if N2 not in tlist:
                    tlist.append(N2)
                    e2 = index
                    nlist[e2] = N2
                    index = index + 1
                else:
                    # find the e1 index by searching the nlist to find the title 
                    for i, t in nlist.items(): 
                        if t == N1: 
                            e2 = i
                            
                edge = [e1,e2] # e1 references e2
                elist.append(edge)
                            
                    
G.add_edges_from(elist)
print(len(G.edges()))
```

```python
Gprime = G
to_remove = []
for e in Gprime.edges(): 
    if e[1] == e[0]:
        to_remove.append(e)
Gprime.remove_edges_from(to_remove)

nodes_to_remove = []
for n in G.nodes(): 
    if G.degree(n) == 1: 
        nodes_to_remove.append(n)
G.remove_nodes_from(nodes_to_remove)
print(G.edges())
```
